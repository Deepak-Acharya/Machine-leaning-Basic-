```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,4,5,scale_units = 'xy' , angles ='xy',scale=1)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Addition of two vectors

```python
vector1 = np.asarray([0,0,2,3])
vector2 = np.asarray([0,0,3,-2])

difference = vector1 - vector2
print(difference)
```

```python
plt.quiver(0,0,2,3,scale_units = 'xy' , angles ='xy',scale=1)
plt.quiver(0,0,3,-2,scale_units = 'xy' , angles ='xy',scale=1,color='r')
plt.quiver(0,0,-1,5,scale_units = 'xy' , angles ='xy',scale=1,color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Multiplying a vectoe with scalar


```python
vector1 = np.asarray([0,0,2,3])

vector2 =  2*vector1

print(vector2)
```

Dot product

```python
a = np.array([2,3])
b = np.array([4,4])
a_dot_b = np.dot(a,b)
print(a_dot_b)
```

```python
c = np.array([5,10,20])
d = np.array([8,32,50])
c_cross_d = np.cross(c,d)
print(c_cross_d)
```

```python
a = np.array([2,5])
v  = np.array([8,-6])

magnitude_v = np.sqrt(sum(v**2))

projection = (np.dot(a,v)/magnitude_v**2)*v

print(projection)
```
