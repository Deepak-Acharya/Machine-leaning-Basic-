```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
```

```python
#create the data from nupy using the 'linspace' and then find out their sin and cosine value and put it in graph
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
#sin Wave
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('Sine value')
plt.title('Sine Wave')
plt.show()
```

```python
plt.plot(x,z)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
language = ['English' , 'French' , 'Spanish', 'Latin' , 'German']
people = [100,50,150,40,70]
ax.bar(language,people)
plt.xlabel('Language')
plt.ylabel('People')
plt.show()
```

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
language = ['English' , 'French' , 'Spanish', 'Latin' , 'German']
people = [100,50,150,40,70]
ax.pie(people,labels = language , autopct = '%1.1f%%')
plt.show()
```

```python
x = np.linspace(0,10,20)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color = 'g')
ax.scatter(x,z,color = 'r')
plt.show()
```

```python
fig3 = plt.figure()
ax = plt.axes(projection = '3d')
z = 20*np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z , c=z ,cmap ='Blues')
plt.show()
```
